```python
import pandas as pd
```

```python
headers=["symboing","normalized-losses","make","fuel-type","aspiration","mun-of-doors","body-style","drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type","num-of-cylinders","engine-size","fuel-system","bore","stroke","compresssion-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"]
ds="imports-85.data"
df=pd.read_csv(ds,sep=",",names=headers) 
```
